#### **II. Analysis with rep**

##### Logic: Here, the participant had to compare pairs of different textures. They had to tell if they perceive the pairs as similar or different (same or different task). In each combinaison, they had to compare the reference, texture 2 on of the other texture

```python
import numpy as np
import os
import matplotlib as plt
import glob
import pandas as pd
import seaborn as sns
```

```python
#import data and check proportion good answer 
path = "C:/Users/Jenna/Documents/isir/haptics/psychophysics/fish_tank/new_texture/comp_with_ref/*.csv"
columns = ["T1","T2","PA","Scale"]  #scale: 1 same -9 different #pas de doublons sauf texture 2 (reference)

for filename in glob.glob(path):
    df = pd.read_csv(filename, sep=",", header=None)
    df.columns = columns
    df["TA"] = np.where(df["T1"] == df["T2"], 1,2) #true answer
    df["C"] = np.where(df["PA"]==df["TA"],1,0) #comparision participant answer/true answer
    display(df)
    
    # proportion of correct answer for the entire experiment
    proportion_correct = ((df["C"] == 1).sum())/32
    print("proportion of correct answer = ", proportion_correct)    
    
```
